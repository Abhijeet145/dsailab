```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("gapminder.csv")

# Set Seaborn style
sns.set_theme(style="whitegrid")

# Filter for most recent year (2007) for simplicity
subset_df = df[df['year'] == 2007]

# 1. Line Plot - GDP per Capita Over Time for India
plt.figure()
india = df[df["country"] == "India"]
sns.lineplot(x="year", y="gdpPercap", data=india)
plt.title("Line Plot: GDP per Capita of India Over Time")

# 2. Scatter Plot - GDP vs Life Expectancy
plt.figure()
sns.scatterplot(x="gdpPercap", y="lifeExp", hue="continent", data=subset_df)
plt.title("Scatter Plot: GDP vs Life Expectancy (2007)")

# 3. Bar Plot - Average Life Expectancy by Continent
plt.figure()
sns.barplot(x="continent", y="lifeExp", data=subset_df)
plt.title("Bar Plot: Avg Life Expectancy by Continent (2007)")

# 4. Pie Chart - Population Distribution by Continent
plt.figure()
continent_pop = subset_df.groupby("continent")["pop"].sum()
plt.pie(continent_pop, labels=continent_pop.index, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart: Population by Continent (2007)")

# 5. Histogram - Life Expectancy
plt.figure()
sns.histplot(subset_df["lifeExp"], bins=20, kde=True)
plt.title("Histogram: Life Expectancy (2007)")

# 6. Box Plot - Life Expectancy by Continent
plt.figure()
sns.boxplot(x="continent", y="lifeExp", data=subset_df)
plt.title("Box Plot: Life Expectancy by Continent (2007)")

# 7. Heatmap - Correlation Matrix
plt.figure()
sns.heatmap(subset_df[["lifeExp", "pop", "gdpPercap"]].corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap: Correlation Matrix (2007)")

# 8. Pair Plot - Numeric Relationships
sns.pairplot(subset_df[["lifeExp", "pop", "gdpPercap"]])
plt.suptitle("Pair Plot: LifeExp, Population, GDP per Capita", y=1.02)

# 9. Count Plot - Number of Countries per Continent
plt.figure()
sns.countplot(x="continent", data=subset_df)
plt.title("Count Plot: Number of Countries per Continent (2007)")

# Show all plots
plt.show()
```
